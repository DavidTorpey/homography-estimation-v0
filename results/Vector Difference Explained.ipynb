```python
import torchvision.transforms as T
from PIL import Image
from skimage.data import astronaut
```

```python
image = Image.fromarray(astronaut())
```

```python
image
```

```python
random_affine = T.RandomAffine(180, (0.25, 0.25), (0.7, 1.3), [-25, 25, -25, 25])
```

```python
affine = random_affine(image)
```

```python
affine
```

```python
image.save('original.png')
affine.save('affine.png')
```
